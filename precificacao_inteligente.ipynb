## MANIPULAÃ‡ÃƒO DE TEXTOS

```python
dados['descricao_local'] = dados['descricao_local'].str.lower() # TRANSFORMANDO COLUNA TEXTUAIS EM LETRAS MINUSCULAS
dados
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex = True) # SUBSTITUINDO CARACTERES ESPECIAS INDESEJADOS
```

```python
dados['descricao_local'] = dados['descricao_local'].str.replace('(?<!\w)-(?!\w)', ' ', regex = True) # REMOVENDO HIFENS FORA DE PALAVRAS COMPOSTAS
```

```python
dados['descricao_local'] = dados['descricao_local'].str.split() # TOKENTIZAÃ‡ÃƒO DA COLUNA
dados.head()
```

```python
dados['comodidades'] = dados['comodidades'].str.replace('\{|}|\"', '', regex = True) # SUBSTITUINDO CARACTERES ESPECIAS INDESEJADOS
```

```python
dados['comodidades'] = dados['comodidades'].str.split(',') # TOKENTIZACAO
```

```python
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.lower()

dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace('[^a-zA-Z0-9\-\']', ' ', regex=True)
dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.replace('(?<!\w)-(?!\w)', '', regex=True)

dados['descricao_vizinhanca'] = dados['descricao_vizinhanca'].str.split()
dados.head()
```
